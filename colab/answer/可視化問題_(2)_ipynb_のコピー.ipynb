## 1問目
使うcsvはhist_data.csvです。下にコードを書いてください。

問1.png

```python
data = pd.read_csv('https://raw.githubusercontent.com/yk4nonome/Basic-Python/master/colab/quizzes/matplotlib_quiz/hist_data.csv')
a =data['aデータ']
b =data['bデータ']
A =a.T
B =b.T
kwargs = dict(histtype ='stepfilled',alpha=0.5,bins=40)
plt.hist(A,**kwargs,label="データa")
plt.hist(B,**kwargs,label="データｂ")
plt.title("aデータとｂデータの温度に関する度数分布")
plt.axis([0,55,0,27])
plt.legend();
plt.xlabel("温度℃")
plt.ylabel("度数")
```

## 2問目
使うcsvはscatter_data.csvです。下にコードを書いてください。

問2.png

```python
data = pd.read_csv('https://raw.githubusercontent.com/yk4nonome/Basic-Python/master/colab/quizzes/matplotlib_quiz/scatter_data.csv')
Class =data['class']
dass0 = (Class ==0)
dass1 = (Class==1)
dass2 = (Class==2)
Length =data['length (cm)']
Width =data['width (cm)']
plt.scatter(Length[dass0],Width[dass0],c="red",label='dass0')
plt.scatter(Length[dass1],Width[dass1],c="blue",label='dass1')
plt.scatter(Length[dass2],Width[dass2],c="green",label='dass2')
plt.axis([4.5,8.0,2.0,4.5])
plt.title('クラス０～２のlengthとwidthに関する散布図')
plt.xlabel('length(cm)')
plt.ylabel('width(cm)')
plt.grid()
plt.legend()
```

## 3問目
使うcsvはline_data.csvです。下に2つの画像を表示するコードを書いてください。

問3(1).png

問3(2).png

```python
data =pd.read_csv('https://raw.githubusercontent.com/yk4nonome/Basic-Python/master/colab/quizzes/matplotlib_quiz/line_data.csv')
term=data['期間']
number=data['件数']
cost=data['費用']
#plt.subplot(2,1,1)
plt.plot(term,number)
plt.rcParams["figure.figsize"] = (2,1)
plt.xlabel('期間')
plt.ylabel('件数')
plt.title('期間ごとの件数の推移')
plt.grid()
plt.xticks(rotation=45)
plt.figure(figsize=[7,7])
#plt.subplots_adjust(hspace=3)
#plt.subplot(2,1,2)
plt.bar(term,cost)
plt.xticks(rotation=45)
plt.rcParams["figure.figsize"] = (2,1)
plt.xlabel('期間')
plt.ylabel('費用（円）')
plt.grid(linestyle='-')
plt.title('期間ごとの費用の推移')
plt.show
```
